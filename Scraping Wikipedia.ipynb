```python
import pandas as pd
from bs4 import BeautifulSoup as soup
from bs4 import NavigableString, Tag
import requests
import csv 

years = range(1985,2017) # have a list of years
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
base_url = "https://en.wikipedia.org/wiki"

df = pd.DataFrame()
# iterate over each year's wiki page to build a bag of words of good/bad things that happened
for year in years:
    print('.', end='')
    url = f"{base_url}/{year}"
    response = requests.get(url)
    if response.status_code != 200:
        print(f"GET failed with response code: {response.status_code}")
        raise
        
    year_soup = soup(response.text, 'html5lib')
    uls = year_soup.find_all('ul')
    # iterate over each month
    for month in months:
        # iterate over all the uls in the hardcoded range of desired uls
        for i, ul in enumerate(uls[16:]):
            if month in ul.get_text():
#                 print(f"I ==== {i}", ul.get_text())
                df = df.append({'year': int(year), 'month': month, 'text': ul.get_text().encode('ascii','ignore').decode()}, ignore_index=True)
# #         bullets = [year.get_text() for year in year_soup.find_all('ul')]
#         print(ms)
df
```

```python
df_csv = df.to_csv()
with open('Project1_sentiment.csv', 'w') as f:
    f.write(df_csv)
```

```python
# Alt approach using the "next_element"
from bs4 import NavigableString, Tag


years = range(1985,2017) # have a list of years
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
base_url = "https://en.wikipedia.org/wiki"

df2 = pd.DataFrame()
# iterate over each year's wiki page to build a bag of words of good/bad things that happened
for year in years:
    url = f"{base_url}/{year}"
    response = requests.get(url)
    if response.status_code != 200:
        print(f"GET failed with response code: {response.status_code}")
        raise
        
    year_soup = soup(response.text, 'html5lib')
    for month in months:
        print(f"GETTING {year}-{month}")
        span = year_soup.find('span', {'id': month})
        for elem in span.parent.findNext('ul'):
            if isinstance(elem, NavigableString):
                continue
            elif isinstance(elem, Tag):
#                 print(elem.get_text())
                df2 = df2.append({'year': int(year), 'month': month, 'text': elem.get_text().encode('ascii','ignore').decode()}, ignore_index=True)
df2
```
